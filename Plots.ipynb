```python
import numpy as np
import pandas as pd
import pylab
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d 
```

```python
X = np.array([[0.18, 0.89],[1.  , 0.26],[0.92, 0.11],[0.07, 0.37],[0.85, 0.16],[0.99, 0.41],[0.87, 0.47]])
Y = np.array([[109.85],[155.72],[137.66],[ 76.17],[139.75],[162.6 ],[151.77]])
```

```python
X
```

```python
ax = plt.axes(projection ='3d')
ax.scatter(X[:,0], X[:,1], Y [:,0])
ax.plot3D(X[:,0], X[:,1], Y [:,0], 'blue') 
ax.set_title('3D line plot') 
plt.show()
```

```python
plt.scatter(X[:,0], X[:,1])
plt.title('2D plot')
plt.xlabel('Feature1')
plt.ylabel('Feature2')
plt.show()

plt.scatter(X[:,0], Y[:,0])
plt.title('2D plot')
plt.xlabel('Feature1')
plt.ylabel('Dependent')
plt.show()

plt.scatter(X[:,1], Y[:,0])
plt.title('2D plot')
plt.xlabel('Feature2')
plt.ylabel('Dependent')
plt.show()
```
